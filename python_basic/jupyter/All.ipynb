```python
nums1 = [1, 2, 3]
nums2 = [1, 2, 3]
nums3 = [1, 2, 3]
```

```python
# 新增
nums1.append(4)
print(nums1)
```

```python
nums2.insert(0, 0)
print(nums2)
```

```python
nums3.extend(range(4,6))
print(nums3)
```

```python
# 删除
result = nums1.pop()
print(nums1, result)
```

```python
nums1.remove(2)
```

```python
print(nums1)
```

```python
nums2.clear()
print(nums2)
```

```python
del nums3[2]
print(nums3)
```

```python
i = 0
while i < len(nums3):
    print(nums3[i])
    i += 1
```

```python
for i in nums3:
    print(i)
```

```python
for i,item in enumerate(nums3):
    print(i, item)
```

```python
for i,item in enumerate(nums3,start=1):
    print(i, item)
```
