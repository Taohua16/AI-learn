```python
import matplotlib.pyplot as plt

def render_latex(text, filename='formula.png'):
    # 创建一个新的图形
    fig = plt.figure(figsize=(6, 4))
    ax = fig.add_axes([0, 0, 1, 1])
    
    # 设置无边框和无刻度
    ax.set_axis_off()
    
    # 使用text函数添加LaTeX文本到图像中
    ax.text(0.5, 0.5, text, size=20, ha="center", va="center",
            bbox=dict(alpha=0.5, edgecolor="b", facecolor="w"))  # 使用usetex=True来启用LaTeX渲染
    
    # 保存图片
    plt.savefig(filename)
    plt.close(fig)
```

```python
# 设置中文字体（推荐使用系统已有的字体）
plt.rcParams['font.sans-serif'] = ['SimHei', 'FangSong', 'Microsoft YaHei', 'DejaVu Sans']
plt.rcParams['axes.unicode_minus'] = False  # 解决负号 '-' 显示为方块的问题
```

```python
latex_text = r"""
因为 $OA = OB$，且 $\angle AOB = 90^\circ$，所以 $\angle OAB = 45^\circ$。
因为 $PC \perp OA$，所以 $\triangle ACD$ 为等腰直角三角形，因此 $AC = CD = 1$。
设该扇形的半径长为$r$，则 $OC = r - 1$，
在 $\mathrm{Rt}\triangle POC$ 中，$\angle PCO = 90^\circ$，$PC = PD + CD = 3$，
所以 $OP^{2} = OC^{2} + PC^{2}$，即
\[ r^{2} = (r - 1)^{2} + 9 \]
解得：$r = 5$。
"""

# 调用render_latex函数
render_latex(latex_text, 'output_formula.png')

print("公式已渲染并保存为 output_formula.png")
```
