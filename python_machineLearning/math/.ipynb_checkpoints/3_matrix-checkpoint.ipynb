```python
import numpy as np
```

```python
A = np.array([[1,2,3],[4,5,6]])
B = np.array([[1,3,7],[5,0,2]])
print(A)
print(A.shape)
```

```python
# 1.转置
print(A.T)
print(A.T.shape)
```

```python
# 2.矩阵相加
print(A + B)
```

```python
# 3.矩阵与标量相乘
print(A * 3) 
```

```python
# 4.矩阵与向量相乘 (列数必须相同)
x = np.array([1,3,4])
print(A * x)
```

```python
# 5.矩阵与矩阵相乘 （对应位置相乘, 哈达玛积）**形状必须一样
print(A * B)
```

```python
# 6.矩阵乘法
# print(A.dot(B))
print(A.dot(B.T))
```

```python
print(A @ B.T)
```

```python
# 7.求逆
C = np.array([[1,2],[3,5]])
c_inv = np.linalg.inv(C)
print(c_inv)
```
