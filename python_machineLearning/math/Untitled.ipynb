```python
import numpy as np
```

```python
x = np.array([2, 5, 8])
print(x)
print(x.shape)
```

```python
# 1.向量转置
print(x.T)
print(x.T.shape)
```

```python
# 2.相加
y = np.array([1, 3, 7])
print(x + y)
```

```python
# 3.向量与标量相乘
print(x * 3)
```

```python
# 4.向量与向量相乘（对应位置元素相乘）
print(x*y)
```

```python
# 5.向量点乘
print(x.dot(y))
print(x @ y)
```

```python
# 6.范数
L0_norm = np.linalg.norm(x,ord=0)
print(L0_norm)
L1_norm = np.linalg.norm(x,ord=1)
print(L1_norm)
L2_norm = np.linalg.norm(x,ord=2)
print(L2_norm)
```
