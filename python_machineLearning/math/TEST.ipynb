```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pylab import mpl
```

```python
    # 定义 y = sinx
x = np.arange(0,6,0.1)
y = np.sin(x)
```

```python
print(y)
```

```python
# 求导
y1 = np.cos(x)
```

```python
plt.plot(x,y1)
plt.xlabel("x")
plt.ylabel("y")
plt.title("y = cos(x)")
plt.show()
```

```python

plt.plot(x,y,label="sin(x)")
plt.plot(x,y1,label="cos(x)",linestyle="--")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
```
