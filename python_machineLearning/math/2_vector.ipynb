```python
import numpy as np
```

```python
# 定义向量
x = np.array([2, 5, 8])
print(x)
print(x.shape)
```

```python
# 1. 向量转置
print(x.T)
print(x.T.shape)
```

```python
# 2. 相加
y = np.array([1, 3, 7])
print(x + y)
```

```python
# 3. 向量与标量相乘
print(x * 3)
```

```python
# 4. 向量与向量相乘（对应位置元素相乘）
print(x * y)
```

```python
# 5. 向量点乘
np.dot(x, y)
print(x.dot(y))
```

```python
print(x @ y)
```

```python
# 6. 计算范数
l0_norm = np.linalg.norm(x, ord=0)
print(l0_norm)
```

```python
l2_norm = np.linalg.norm(x, ord=2)
print(l2_norm)
```
